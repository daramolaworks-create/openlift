# OpenLift Geo-Lift Demo

This notebook runs the bundled geo-lift example and prints the decision-ready output.

```python
from pathlib import Path
from openlift.core.pipeline import run_geo_lift

candidates = [
    Path('examples/geo_lift_basic/experiment.yaml'),
    Path('../geo_lift_basic/experiment.yaml'),
]
config_path = next(path for path in candidates if path.exists())
results = run_geo_lift(str(config_path))
results
```

```python
metrics = results['metrics']
decision = results.get('decision', {})

print(f"Incremental outcome: {metrics['incremental_outcome_mean']:.1f}")
print(f"Lift: {metrics['lift_pct_mean']:.1%}")
print(f"P(lift > 0): {metrics['p_positive']:.1%}")
print(f"Decision: {decision.get('decision')}")
print(f"Next action: {decision.get('next_action')}")
```
